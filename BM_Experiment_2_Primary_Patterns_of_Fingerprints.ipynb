```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
```

```python
fingerprint = cv2.imread("/content/drive/MyDrive/NMIMS/Biometrics/Fingerprint.jpg")
fingerprint_res = cv2.resize(fingerprint,(0,0), fx=0.1,fy=0.1)
```

```python
cv2_imshow(fingerprint_res)
```

```python
fingerprint_gray = cv2.imread("/content/drive/MyDrive/NMIMS/Biometrics/Fingerprint.jpg",0)
fpinit = cv2.resize(fingerprint_gray,(0,0), fx=0.1,fy=0.1)
cv2_imshow(fpinit)
```

```python
fp = fpinit[150:320, 100:250]
cv2.rotate(fp, cv2.ROTATE_180)
cv2_imshow(fp)
```

```python
ret,thresh = cv2.threshold(fp,127,255,cv2.THRESH_BINARY)
cv2_imshow(thresh)
```

```python
adaptive_thresh = cv2.adaptiveThreshold(
    fp,
    255,
    cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
    cv2.THRESH_BINARY_INV,
    11,
    2
)
cv2_imshow(adaptive_thresh)
```
